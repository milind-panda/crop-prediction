```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix
```

```python
dataset = pd.read_csv("/content/Crop_recommendation.csv")
features = dataset.iloc[:, 0:7].values
labels = dataset.iloc[:, 7].values
```

```python
ratio_train = 0.8
ratio_val = 0.1
ratio_test = 0.1
```

```python
x_remaining, x_test, y_remaining, y_test = train_test_split(features, labels, test_size=ratio_test, random_state=10)
ratio_remaining = 1 - ratio_test
ratio_val_adjusted = ratio_val / ratio_remaining
x_train, x_val, y_train, y_val = train_test_split(x_remaining, y_remaining, test_size=ratio_val_adjusted, random_state=10)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
x_test = scaler.transform(x_test)
model = svm.SVC()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
def predict_crop():
    print("\nEnter the following soil and climate parameters:")
    N = float(input("Nitrogen (N): "))
    P = float(input("Phosphorus (P): "))
    K = float(input("Potassium (K): "))
    temperature = float(input("Temperature (°C): "))
    humidity = float(input("Humidity (%): "))
    ph = float(input("pH value: "))
    rainfall = float(input("Rainfall (mm): "))

    input_features = np.array([[N, P, K, temperature, humidity, ph, rainfall]])

    input_scaled = scaler.transform(input_features)

    prediction = model.predict(input_scaled)
    print("\n Recommended Crop:", prediction[0])

predict_crop()
```
